```python
import matplotlib
matplotlib.__version__
```

```python
import matplotlib.pyplot as plt
```

```python
savings=[100,200,400,900,500,400,300,700,1000]
months=['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep']
```

```python
plt.plot(months,savings)
plt.xlabel('Months')
plt.ylabel('Savings Amount ($)')
plt.title('Savings by Month')
```
